```python
import numpy as np
from scipy import stats
import pandas as pd
from scipy.stats import chi2_contingency
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.weightstats import ztest

```

```python
A = [33,36,39,45,40,37,40,41]
B = [40,41,55,60,52,55,45,49]
C=  [39,43,51,67,50,45,40,52]
```

```python
print(np.mean(A))
print(np.mean(B))
print(np.mean(C))
```

```python
group1 = np.array(A)
group2 = np.array(B)
group3 = np.array(C)
```

```python
f_stat,p_value = stats.f_oneway(group1,group2,group3)
```

```python
print(f"f_stat = {f_stat} , p_value = {p_value}")
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("One group mean is different")
else:
    print("No significant difference")
```

```python
data = A+B+C
```

```python
labels = (
    ['group1']*len(A) + 
    ['group2']*len(B) +
    ['group3']*len(C)
)
```

```python
df = pd.DataFrame(
    {
    "values":data,
    "groups":labels
    }
)
```

```python
df
```

```python
tukey=pairwise_tukeyhsd(endog=df["values"],
                        groups=df["groups"],
                        alpha = 0.05
                       )
```

```python
# one sample T-test : when you want to compare sample mean vs population mean
```

```python
data = [45,48,50,52,47]
mu_0 = 55 
# mu_0 is population mean
```

```python
t_stat,p_value=stats.ttest_1samp(data,mu_0)
```

```python
print(f"t_stat={t_stat} , p_value={p_value}")
```

```python
if p_value < 0.05:
    print("significant difference")
else:
    print("no significant difference")
```

```python
# Anova test : used to compare 3 or more groups 
```

```python
A = [40,42,41]
B = [55,57,56]
C = [48,50,47]
```

```python
f_stat,p_value=stats.f_oneway(A,B,C)
```

```python
print(f"t_stat={t_stat} , p_value={p_value}")
```

```python
if p_value < 0.05:
    print("significant difference")
else:
    print("no significant difference")
```

```python
# tukey test apply on above to check which group differs from which
```

```python
# combine all data into one list
data = A+ B + C
```

```python
# create a group labels
groups = (["A"]*len(A) + 
          ["B"]*len(B) + 
          ["C"]*len(C)
         )
```

```python
groups
```

```python
tukey = pairwise_tukeyhsd(endog=data,
                          groups=groups,
                          alpha = 0.05)
```

```python
print(tukey)
```

```python
# conclusion : All three groups are significantly different from each other.
# There is no pair that is similar.
```

```python
# Z-Test: used when sample size is larger and population std deviation is known
```

```python
data = [52,54,51,53,55,56,50,52,54,53,
        51,52,53,55,54,52,53,51,56,57,
        54,53,52,55,54,53,52,51,56,55,
        53,52,54,55,53,52,51,54,55,56]
mu_0 = 50
```

```python
z_stat,p_value = ztest(data,value = mu_0)
```

```python
print("z-staticstic",z_stat)
print("p_value",p_value)
```

```python
if p_value < 0.05:
    print("Significant Difference (Reject H0)")
else:
    print("No Significant Difference (Accept H0)")
```

```python
# outliers: A value which is too far away from the rest of the data
```

```python
# Z-score method to detect ouliers
# z-score tells How many standard deviations a data point is away from the mean
```

```python
data = [10,12,11,13,140,15,16,17,18,19,100,120]
```

```python
z_scores = stats.zscore(data)
```

```python
print(z_scores)
```

```python
# find outliers
outliers = []
for i in range(len(z_scores)):
    if abs(z_scores[i]) > 2:
        outliers.append(data[i])
print("outliers",outliers)
```

```python
# Modified z-score: best for multiple outliers 
# it uses median instead of mean 
# MAD(median absolute deviation)
```

```python
data = np.array([10,12,11,13,14,15,16,17,18,19,100,120])
```
