```python
import numpy as np
from scipy import stats
import pandas as pd
from scipy.stats import chi2_contingency
from statsmodels.stats.multicomp import pairwise_tukeyhsd

```

```python
A = [33,36,39,45,40,37,40,41]
B = [40,41,55,60,52,55,45,49]
C=  [39,43,51,67,50,45,40,52]
```

```python
print(np.mean(A))
print(np.mean(B))
print(np.mean(C))
```

```python
group1 = np.array(A)
group2 = np.array(B)
group3 = np.array(C)
```

```python
f_stat,p_value = stats.f_oneway(group1,group2,group3)
```

```python
print(f"f_stat = {f_stat} , p_value = {p_value}")
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("One group mean is different")
else:
    print("No significant difference")
```

```python
data = A+B+C
```

```python
labels = (
    ['group1']*len(A) + 
    ['group2']*len(B) +
    ['group3']*len(C)
)
```

```python
df = pd.DataFrame(
    {
    "values":data,
    "groups":labels
    }
)
```

```python
df
```

```python
tukey=pairwise_tukeyhsd(endog=df["values"],
                        groups=df["groups"],
                        alpha = 0.05
                       )
```

```python
print(tukey)
```

```python
print(np.mean(A))
print(np.mean(B))
print(np.mean(C))
```

```python
# Co-relation
```

```python
x = np.array([1,2,4,6,7,9])
y = np.array([50,55,67,75,80,89])
```

```python
corr, p_value = stats.pearsonr(x,y)
```

```python
print(round(corr,3))
print(p_value)
```

```python
if p_value < 0.05:
    print("Significant linear relationship exist")
else:
    print("No significant linear relationship exist")
```

```python
# Chi square test
```

```python
table  = np.array([
    [10,20,30],
    [6,9,17]
])
```

```python
chi2,p,dof,expected= chi2_contingency(table)
```

```python
print("chi value",round(chi2,3))
print("p_value",round(p,3))
print("degree of freedom",dof)
```

```python
print(np.round(expected,2))
```

```python
alpha = 0.05
```

```python
if p_value < 0.05:
    print("Significant relationship exist")
else:
    print("No significant relationship exist")
```

```python
n=table.sum()
k = min(table.shape)
```

```python
cramers_v = np.sqrt(chi2/(n*(k-1)))
print(cramers_v)
```

```python
if cramers_v < 0.1:
    print("Very Weak")
elif cramers_v < 0.3:
    print("Weak")
elif cramers_v < 0.5:
    print("Moderate")
else:
    print("Strong")
```
