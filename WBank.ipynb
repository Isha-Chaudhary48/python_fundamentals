```python
import re
import secrets
```

```python
# Authentication 
class Auth:
    def check_email(self,email):
        if not email.endswith(('@gmail.com','@yahoo.com')):
            return False,"Invalid email , enter valid email"
        else:
            return True,"Email successfully checked"
    def check_password(self,password):
        if ' ' in password:
            return False,"Space are not allowed in password"
        elif len(password) < 6:
            return False,"Password length should be greater than 6"
        elif not re.search(r"\d",password):
            return False,"Password should include alteast one digit"
        elif not re.search(r"[a-z]",password):
            return False,"Password should include atleast one lower case letter"
        elif not re.search(r"[A-Z]",password):
            return False,"Password should include atleast one upper case letter"
        elif not re.search(r"[!@#$$%^&*()_]",password):
            return False,"Password should include atleast one special character"
        else:
            return True , "Strong Password"
            
        
    
```

```python
class User:
    def __init__(self,bank,name,email,inputPassword):
        self.bank = bank
        self.name= name
        self.email = email
        self.password = inputPassword
    def userDataSave(self):
        self.bank.userDataList[self.email] = {
            "name":self.name,
            "password":self.password
        }
        print("\tUser info saved successfully")
        
        
```

```python
# Bank Account Creation
class Account:
    def __init__(self,bank,email,balance = 0):
        self.email = email
        self.bank = bank
        self.balance = balance
        self.transactionList = [f"Opening balance : {balance}"]
    def createAccount(self):
        accountNo = "".join(secrets.choice('0123456789') for _ in range(8)) 
        print(f"\tYour Bank Account no is {accountNo}")
        self.bank.userAccountList[accountNo] = {           
            "email" : self.email,
            "balance" : self.balance,
            "transactionList" : self.transactionList
        }
        print("\tAccount successfully created ")
        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")       
    
```

```python
# Bank
class WBank:
    def __init__(self):
        self.userDataList={}
        self.userAccountList={}       
    def menu(self):
        self.auth = Auth()
        while True:
            print("----------------------------------------------------------------")
            print("\tWelcome to WBank")
            print("\tHow can we help you")
            print("\tBelow are the options you can choose any service you want")
            print("\t1. Create an Bank Account")
            print("\t2. Login to existing Account")
            print("\t3. Exit")
            print("----------------------------------------------------------------")
            choice = int(input("Enter any choice from above: "))
            match choice:
                case 1:
                    name = input("Enter your name: ")
                    max_attempts = 3
                    attempt = 0
                    while True:
                        email = input("Enter your email id: ")
                        if email in self.userDataList:
                            print("Email already registered please try again with another email")
                            return
                            
                        emailStatus,emailMsg= self.auth.check_email(email)
                        if emailStatus:
                            print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                            print(emailMsg)
                            print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                            break
                        else:
                            attempt +=1
                            print(emailMsg)
                            if attempt == max_attempts:
                                print("Too many attempts try again later...")
                                return
                    while True:
                        inputPassword= input("Enter password: ")
                        passwordStatus,passwordMsg = self.auth.check_password(inputPassword)
                        if passwordStatus:
                            print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                            print("\t",passwordMsg)
                            user1 = User(self,name,email,inputPassword)
                            user1.userDataSave()
                            bankAccount1 = Account(self,email)
                            bankAccount1.createAccount()
                            break
                        else:
                            print(passwordMsg)                                         
                case 2:
                    accountNo = input("Enter your account no: ")
                    if accountNo not in self.userAccountList:
                        print("Account doesnot exist. Enter valid account no.")
                        return
                        
                    userEmail = self.userAccountList[accountNo]['email']
                    max_attempts = 3
                    attempt = 0
                    while True: 
                        userPassword = input("Enter password: ")
                        if userPassword != self.userDataList[userEmail]['password']:                        
                            print("Enter valid password")
                            attempt +=1
                            if attempt == max_attempts:
                                print("Too many attempts try again later...")
                                return
                        else:
                            while True:
                                print("\t1. Deposit money")
                                print("\t2. Withdraw money")
                                print("\t3. Bank to Bank transfer")
                                print("\t4. Check Transaction History")
                                print("\t5. Check Bank Balance")
                                print("\t6. Exit") 
                                choice = int(input("Enter any choice from above: "))
                                match choice:
                                    
                                    case 1:
                                            amount = int(input("Enter amount you want to deposit: "))
                                            self.userAccountList[accountNo]['balance'] += amount
                                            self.userAccountList[accountNo]['transactionList'].append(f"Amount Deposited : {amount}")
                                            print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                            print(f"\tYou have sucessfully deposited {amount}")
                                            print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                    case 2:
                                                amount = int(input("Enter amount you want to withdraw: "))
                                                if self.userAccountList[accountNo]['balance'] < amount:
                                                    print('\n')
                                                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                                    print("\tInsufficent Bank Balance")
                                                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                                    print('\n')
                                                else:
                                                    self.userAccountList[accountNo]['balance'] -= amount
                                                    print('\n')
                                                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                                    self.userAccountList[accountNo]['transactionList'].append(f"Amount withdraw : {amount}")
                                                    print(f"\tYou have successfully withdrawl amount : {amount}")
                                                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                                    print('\n')
                                    case 3:
                                        receiverAccountNo = input("Enter account number you want to send money: ")
                                        if receiverAccountNo in self.userAccountList:
                                            amountSend = int(input("Enter amount you want to send"))
                                            if amountSend > self.userAccountList[accountNo]['balance']:
                                                print("Insufficent balance")
                                                return
                                                
                                            else:
                                                self.userAccountList[accountNo]['balance'] -= amountSend
                                                self.userAccountList[accountNo]['transactionList'].append(f"sended {amountSend} Rs. to account number {receiverAccountNo}")
                                                self.userAccountList[receiverAccountNo]['balance'] += amountSend
                                                self.userAccountList[receiverAccountNo]['transactionList'].append(f"Received {amountSend} Rs. from account number {accountNo}")
                                                print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                                print("\tPayment successfull")
                                                print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        else:
                                            print("Receiver doesn't exist")
                                            return
                                                       
                                    case 4:
                                        transactionHistory = self.userAccountList[accountNo]['transactionList']
                                        print('\n')
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print("\tYour Transaction History")
                                        print("\n")
                                        for i in transactionHistory:
                                            print("\t-" , i)
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print('\n')
                                    case 5:
                                        print('\n')
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print("\tTotal Bank Balance = ", self.userAccountList[accountNo]['balance'])
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print('\n')
                                            
                                        
                                    case 6:
                                        print('\n')
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print("\tSuccessfully exited...")
                                        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                                        print('\n')
                                        return
                                
                                    case _:
                                        print("Invalid choice")
                case 3:
                    print('\n')
                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                    print("\tSuccessfully exited")
                    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
                    print('\n')
                    break 
                    
                                    
                case _:
                    print("Invalid choice")
```
