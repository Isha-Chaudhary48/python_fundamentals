```python
# Match just like a switch in cpp
choice = int(input())
match choice:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Invalid")
```

```python
num1 = int(input("Enter num1"))
num2 = int(input("Enter num2"))
op = input("Operator")
match op:
    case '+':
        print(num1 + num2)
    case '-':
        print(num1 - num2)
    case '*':
        print(num1 * num2)
    case '/':
        print(num1 / num2)
    case '%':
        print(num1 % num2)
    case _:
        print("Invalid operator")
        
```

```python
day = input("Enter day")
match day:
    case 'sat' | 'sun':
        print("Weekend")
    case 'mon' | 'tue' | 'wed' | 'thurs' | 'fri':
        print("Weekday")
    case _:
        print("Invalid Input")
    
```
