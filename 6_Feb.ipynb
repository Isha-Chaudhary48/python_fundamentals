```python
# usecase to read like this 1. do not read full file at one load line by line means less load on memory 2. donot need to close the file automatic close
with open('practice.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# any no of lines we can append closes automatically
with open('practice.txt','a')as file:
    a = int(input("Enter no. of lines you want to append"))
    for i in range(a):
        data = input("Enter the data")
        file.write(data +'\n')
        
```

```python
# zip function use to access more than one list
name = ["Isha","Shweta","Roy"]
marks = [78,89,76]
```

```python
with open('students.txt','w') as file:
    for name, marks in zip(name,marks):
        file.write(f"{name},{marks}\n")
```

```python
with open('students.txt','r') as file:
    read = file.readlines()
    print(read)
```
