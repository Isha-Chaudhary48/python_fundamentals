```python
class A:
    def a(self):
        print("This is A")
class B(A):
    def b(self):
        print("This is B")
class C(B):
    def c(self):
        print("This is C")
        
```

```python
class AA:
    def aa(self):
        print("This is AA")
class BB(AA):
    def bb(self):
        print("This is BB")
class CC(BB):
    def cc(self):
        print("This is CC")
```

```python
class ABC(CC,C):
    def func(self):
        print("Final ABC func")
```

```python
ob1 = ABC()
```

```python
ob1.cc()
ob1.aa()
ob1.c()
```

```python
# To avoid above we use args as the cond is only about amt of arguments
class Calc:
    def __init__(self):
        self.a = 0
    def add(self,*args):
        for i in args:
            self.a += i
        print(self.a)
        
        
```

```python
ob = Calc()
```

```python
ob.add(3,4,5,4)
```

```python
class BankAccount:
    def __init__(self,name,balance = 0):
        self.name = name
        self.balance = balance
        self.transactions=[]
        self.transactions.append(f"Name : {name}")
        self.transactions.append(f"opening balance { balance}")
        print("Welcome to our W Bank")
    def deposit(self,amount):
        if amount > 0:
            self.balance += amount
            print("Total amount now in your bank is " , self.balance)
            self.transactions.append(f"Deposited {amount}, balance = {self.balance}")
        else:
            print("Amount must be in positive")
    def withdraw(self,amount):
        if amount  <= self.balance:
            self.balance -= amount
            print( f"Rs. { amount} successfully withraw amount left = {self.balance} " )
            self.transactions.append(f"withdraw {amount}, balance = {self.balance}")
        else:
            print("Insufficient balance")
            self.transactions(f" Failed to withdraw amount {amount}, balance = {self.balance}")
    def viewBalance(self):
        print(f"name = {self.name} and balance = {self.balance}")
    def Showtransactions(self):
        for i in self.transactions:
            print(i)
        
            
            
```

```python
ob = BankAccount("Isha",500)
```

```python
ob.deposit(500)
```

```python
ob.Showtransactions()
```

```python
import re
import random
import secrets
```

```python
bankUserList={}
```

```python
def bankAccount():
    user = input("Enter user name")
    password = input("Enter strong password")
    if len(password) < 6:
        print("password should include atleast 6 characters")
    elif not re.search(r"[A-Z]",password):
        print("password should include upper case letters")
    elif not re.search(r"[a-z]",password):
        print("password should include lower case letters")
    elif not re.search(r"\d",password):
        print("password should include digit")
    elif not re.search(r"[!@#$%^&*()]",password):
        print("password should include special characters")
    else:
        print("Strong password")
        accountNo = ''.join(secrets.choice('0123456789') for _ in range(8))
        bankUserList[accountNo] = {"user": user, "password": password}
bankAccount()
```
