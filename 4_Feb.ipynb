```python
# if want to find one or more digits use findall it retursn the list of all matches 
```

```python
message = "i got 77 in this and 89 in that and you know 54 in maths"
res = re.findall(r"\d+",message)
print(res)
```

```python
total = 0
for i in res:
    val = int(i)
    total += val
print(total)
```

```python
# sub : substitute used to replace the parts of string that matches the regex pattern with smt else
# re.sub(pattern,replacement,string)
```

```python
msg = re.sub(r"\d","*",message)
# strings are immutable so cannot be changed we need to assign variable and then print
msg
```

```python
#match: only matches the pattern at the beginning of the string whereas search matches in whole string
msg = '1233hhi iam d454 and 34343zuhehe'
res = re.match(r"\d+",msg)
print(res.group())
```

```python
#split : it splits the string into words 
#/s+ matches one or more white space character
msg = re.split(r"\s+",message)
print(msg)
```

```python
st1 = " hello h\nmy\tname is\tisha hh\nand how\tare you"
# \t and \n is used to split strings and give tab and newline but if we print it as it is it will give list 
# if printed via for loop then actual result will be shown up
```

```python
msg = re.split(r"\t+",st1)
print(msg)

```

```python
msg = re.split(r"\n+",st1)
print(msg)
```

```python
st2 = 'hello\nworld'
msg = re.split(r"\n",st2)
print(msg)
for i in msg:
    print(i)
```

```python
# r = raw string Don’t treat backslashes as escape characters.
# ^ start of string, $ end of string 
# ?= must contain
# .* any character
password=input("Enter password")
pattern = r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[!@#$%^&*()])[A-Za-z\d!@#$%^&*()]{8,}$"
if re.match(pattern,password):
    print("strong password")
else:
    print("Weak password")
```

```python
# Lambda function This creates an anonymous function a function without name
```

```python
# normal function 
def add(a,b):
    print(a+b)
add(4,5)
```

```python
add1 = lambda a,b:(a+b)
a = add1(2,4)
a
```

```python
greater = lambda a , b: a if (a > b) else b
greater(9,8)
```

```python
is_even = lambda a : "even" if a % 2 == 0 else "odd"
is_even(8)
```

```python
# map :map() applies a function to every element of a list
def sqr(x):
    return x*x
sqr(5)
```

```python
nums = [2,3,5,7,8]
list(map(sqr,nums))
```

```python
# map mainly used with lambda functions
val = list(map(lambda x:x*x,nums))
print(val)
```

```python
find = list(map(lambda x : "even" if x%2==0 else "odd",nums))
find
```

```python
# filter : only returns element from the list that satisfies the codition
s = list(filter(lambda x : x % 2== 1 ,nums))
s
         
```

```python
marks=[23,30,35,40]
find = list(filter(lambda x : x>30 , marks))
find
```
