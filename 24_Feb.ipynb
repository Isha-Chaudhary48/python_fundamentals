```python
import numpy as np
from scipy import stats
```

```python
data = [22,20,19,24,23,21]
mu_0 = 20
```

```python
n = len(data)
mean = np.mean(data)
std = np.std(data,ddof=1)
# Delta Degrees of Freedom : Sample Standard Deviation When you have SOME data N − 1
```

```python
print(f"number of elements {n} , mean :{mean} , std : {std}")
```

```python
t_stat = (mean-mu_0)/(std/np.sqrt(n))
p_value = 2*(1-stats.t.cdf(abs(t_stat),df=n-1))
```

```python
print(f" t_stat is {t_stat} , p_value is {p_value} ")
```

```python
confidence = 0.95
t_critical = stats.t.ppf((1+confidence)/2, df=n-1)
margin_of_error = t_critical*(std/np.sqrt(n))
```

```python
print(t_critical,margin_of_error)
```

```python
lower = mean - margin_of_error
upper = mean + margin_of_error
```

```python
print(f"lower = {lower}, upper = {upper}")
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("significant difference")
else:
    print("No significant difference")
```

```python
print(p_value,alpha)
```

```python
print(np.mean(group1))
print(np.mean(group2))
```

```python
group1= np.array(group1)
group2 = np.array(group2)
```

```python
n1=len(group1)
n2 = len(group2)
```

```python
mean1 = np.mean(group1)
mean2 = np.mean(group2)
```

```python
std1 = np.std(group1,ddof=1)
std2= np.std(group2,ddof = 1)
```

```python
sp = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2)/(n1+n2 -2))
```

```python
se = sp*np.sqrt(1/n1+1/n2)
```

```python
t_stat=(mean1-mean2)/se
```

```python
df = n1+n2 -2
```

```python
p_value = 2*(1-stats.t.cdf(abs(t_stat),df=df))
```

```python
confidence = 0.95
```

```python
t_critical = stats.t.ppf((1+confidence)/2,df=df)
```

```python
margin_of_error = t_critical * se
```

```python
lower = (mean1-mean2)-margin_of_error
```

```python
upper = (mean1-mean2)+margin_of_error
```

```python
print(f"mean difference = {round(mean1-mean2,2)} ")
print(f"range is {round(lower,3)} to {round(upper,3)} ")
print("p_value",p_value)
print("standard deviation",std1, std2)
```

```python
if p_value < 0.05:
    print("Significant difference")
else:
    print("No significant difference")
```
