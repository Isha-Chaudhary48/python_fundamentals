```python
import random
secret = random.randint(1,100)
attempt = 0;
while True:
    num = int(input())
    attempt += 1
    if num == secret:
        print("You won in ",attempt,"attempts")
        break
    elif num > secret:
        print("Too high")
    else:
        print("Too low")
    
```

```python
import random
secret = random.randint(1,100)
num = -1
attempt = 0
while num != secret:
    num = int(input())
    attempt += 1
    difference = abs(secret - num)
    if difference > 20:
        print("❄️ Very Cold")
    elif difference  <= 5:
        print("🔥 Very Hot")
    elif difference  <= 10:
        print("🌡️ Warm")
    elif difference  <= 20:
        print("😐 Cold")
print("correct answer in " , attempt,"attempts") 
    
        
        
    
    
```
