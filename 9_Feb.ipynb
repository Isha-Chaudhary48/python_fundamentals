```python
from abc import ABC, abstractmethod
```

```python
class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
    @abstractmethod
    def details(self):
        pass
```

```python
class employees(employee):
    def __init__(self):
        self.name = input("Enter your name")
        self.salary = int(input("Enter your salary"))
        self.pf = self.salary*0.20
        self.calculate_salary()
        self.details()
    def calculate_salary(self):
        self.net_salary = self.salary + self.pf
        print("net salary is ",self.net_salary)
    def details(self):
        print(f"name of employee is {self.name} and its salary is {self.net_salary}")
        
```

```python
user1 = employees()
```

```python
user1.calculate_salary()
```

```python
try:
    a = int(input("enter a"))
    b = int(input("enter b"))
    val = a/ b
    print(val)
except(ZeroDivisionError):
    print("number cannot be divided by 0")
except(ValueError):
    print("Enter number only")
finally:
    print("program ran successfully")
```
