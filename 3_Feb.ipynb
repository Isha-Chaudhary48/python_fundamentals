```python
def calc(*args):
    sum = 0
    for i in args:
        sum += i
    return sum
```

```python
calc(3,5,6,21,6,3,6,7,9)
```

```python
def average(*args):
    sum = 0
    for i in args:
        sum+=i
    length = len(args)
    avg = sum/length
    return avg
    
    
```

```python
average(1,2,3,4,5)
```
